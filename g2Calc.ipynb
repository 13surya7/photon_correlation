```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt

chL = np.array([5,6,11,12])
t = np.array([0,0,0,0])

ctw = 6
delay = np.arange(-25,25,1)*ctw 
dA = np.transpose(np.tile(delay, (3,1)))
cA = np.array([ctw, ctw, ctw])

cM = np.zeros(np.shape(dA))

tA = []
file = open('converted/spdc_100s.txt', 'r')
file.readline()
for i in range(10**5):
    ch, tt = list(map(int, file.readline().split()))
    t = (ch != chL)*t + (ch == chL) * tt
    dt= (t - t[0])[1:]
    cM = cM + (np.abs(dA + dt) <= cA)
    tA.append(tt)
    
for i in range(3):
    plt.plot(delay, cM[:,i])
    
tA = np.array(tA)
tD = np.diff(tA)
plt.show()
plt.hist(tD[tD<10**4], 101)
plt.show()
```
